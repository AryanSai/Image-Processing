```python
import cv2
import matplotlib.pyplot as plt

# specify the image path
im_bgr = cv2.imread('/home/dmacs/Desktop/MTech/104P/pic1.png')

plt.imshow(im_bgr)
plt.axis('off')
```

```python
# Convert the image from BGR to RGB
im_rgb = cv2.cvtColor(im_bgr, cv2.COLOR_BGR2RGB)

plt.imshow(im_rgb)
plt.axis('off') # turn off the axis
```

```python
# convert the image to grayscale
im_gray = cv2.cvtColor(im_bgr, cv2.COLOR_BGR2GRAY)

plt.imshow(im_gray, cmap = 'gray')
plt.axis('off')
```

```python
# resize an image to (width, height)
im_res = cv2.resize(im_rgb, (300,300))

plt.imshow(im_res)
plt.axis('off')
```

```python
# rotate an image
im_rot = cv2.rotate(im_rgb, cv2.ROTATE_90_COUNTERCLOCKWISE)

plt.imshow(im_rot)
plt.axis('off')
```

```python
# set specific parts of the image to 0
for i in range(20,181):
    for j in range(20,181):
        im_bgr[i][j] = [0,0,0]

plt.imshow(im_bgr)
plt.axis('off')
```

```python
# you can set it to any other color as well
for i in range(20,181):
    for j in range(20,181):
        im_bgr[i][j] = [10,50,67]

plt.imshow(im_bgr)
plt.axis('off')
```

```python
# saving an image
cv2.imwrite('/home/dmacs/Desktop/MTech/104P/img_bgr.jpg',im_bgr)
```

```python
#NEGATIVE OF A gray IMAGE
bgr = cv2.imread("pic1.png")
original=cv2.cvtColor(bgr,cv2.COLOR_BGR2GRAY)

negative = 255 - original
plt.imshow(negative,cmap='gray')
```

```python
#NEGATIVE OF AN RGB IMAGE
bgr = cv2.imread("pic1.png")
original=cv2.cvtColor(bgr,cv2.COLOR_BGR2RGB)

negative = 255 - original
plt.imshow(negative)
```

```python
#gaussian blur
blurred_image = cv2.GaussianBlur(original, (11, 11), 0)

plt.imshow(blurred_image)
plt.show()

blurred_image2 = cv2.GaussianBlur(original, (19, 19), 0)
plt.imshow(blurred_image2)
plt.show()

plt.imshow(cv2.subtract(blurred_image,blurred_image2))
plt.show()
```

```python
#flipping image vertical
flipped_img=cv2.flip(original,1)
plt.imshow(flipped_img)
```

```python
#flipping image horizontal
flipped_img=cv2.flip(original,0)
plt.imshow(flipped_img)
```

```python
#inverting an image
invert=cv2.bitwise_not(original)

plt.imshow(original)
plt.show()
plt.imshow(invert)
plt.show()
```

```python
#blur an image

blur=cv2.blur(original,(12,12))
plt.imshow(blur)
```

```python
#edge of an image
img_edge = cv2.Canny(original, 290, 290)
plt.imshow(img_edge)
```

```python
#output only the blue channel
dummy=cv2.imread('pic1.png')
dummy=cv2.cvtColor(dummy,cv2.COLOR_BGR2RGB)
for i in range(0,413):
    for j in range(0,413):
        dummy[i][j][1]=0
        dummy[i][j][0]=0

plt.imshow(dummy)        
```
