```python
# Revision
import cv2
import matplotlib.pyplot as plt

# Ensure both images have same dimension (otherwise resize)
im1 = cv2.imread('/content/Untitled design (1).png')
im2 = cv2.imread('/content/Untitled design.png')
im3 = cv2.imread('/content/3.jpg')
im3=cv2.resize(im3,(413,413))
plt.imshow(im3)

# Union (OR) of images
union = cv2.max(im1, im2)

# Intersection (AND) of images
intersection = cv2.min(im1, im2)

# Difference (XOR) of images
difference = cv2.absdiff(im1, im2)

# Complement of the image
complement = cv2.bitwise_not(im1)

# Display the results
plt.subplot(221)
plt.imshow(union)
plt.title('Union')
plt.axis('off')

plt.subplot(222)
plt.imshow(intersection)
plt.title('Intersection')
plt.axis('off')

plt.subplot(223)
plt.imshow(difference)
plt.title('Difference')
plt.axis('off')

plt.subplot(224)
plt.imshow(complement)
plt.title('Complement')
plt.axis('off')
```

```python
#verify
# distributivity
#a u (b n c) = (a u b) n (a u c)
bnc=cv2.min(im2,im3)
aubnc=cv2.max(im1,bnc)

aub=cv2.max(im1,im2)
auc=cv2.max(im1,im3)
aubnauc=cv2.min(aub,auc)

plt.subplot(131)
plt.imshow(aubnc)
plt.axis('off')

plt.subplot(132)
plt.imshow(aubnauc)
plt.axis('off')

diff = cv2.subtract(aubnc,aubnauc)

plt.subplot(133)
plt.imshow(diff)
plt.axis('off')
```

```python
# distributivity with intersection
#a n (b u c) = (a n b) u (a n c)
buc=cv2.max(im2,im3)
anbuc=cv2.min(im1,buc)

anb=cv2.min(im1,im2)
anc=cv2.min(im1,im3)
anbuanc=cv2.min(anb,anc)

plt.subplot(131)
plt.imshow(aubnc)
plt.axis('off')

plt.subplot(132)
plt.imshow(aubnauc)
plt.axis('off')

diff = cv2.subtract(anbuc,anbuanc)

plt.subplot(133)
plt.imshow(diff)
plt.axis('off')
```

```python
# assosciativity with union
# au(buc)=(aub)uc
buc=cv2.max(im2,im3)
aubuc1=cv2.max(im1,buc)

aubuc2=cv2.max(aub,im3)

plt.subplot(131)
plt.imshow(aubuc1)
plt.axis('off')

plt.subplot(132)
plt.imshow(aubuc2)
plt.axis('off')

diff = cv2.subtract(aubuc1,aubuc2)

plt.subplot(133)
plt.imshow(diff)
plt.axis('off')
```

```python
# assosciativity with intersection
# an(bnc)=(anb)nc
anbnc1=cv2.min(im1,bnc)

anb=cv2.min(im1,im2)
anbnc2=cv2.min(anb,im3)

plt.subplot(131)
plt.imshow(anbnc1)
plt.axis('off')

plt.subplot(132)
plt.imshow(anbnc2)
plt.axis('off')

diff = cv2.subtract(anbnc1,anbnc2)

plt.subplot(133)
plt.imshow(diff)
plt.axis('off')
```

```python
# commutativity wrt union
aub=cv2.max(im1,im2)
bua=cv2.max(im2,im1)

plt.subplot(131)
plt.imshow(aub)
plt.axis('off')

plt.subplot(132)
plt.imshow(bua)
plt.axis('off')

diff = cv2.subtract(aub,bua)

plt.subplot(133)
plt.imshow(diff)
plt.axis('off')
```

```python
# commutativity wrt intersection
anb=cv2.min(im1,im2)
bna=cv2.min(im2,im1)

plt.subplot(131)
plt.imshow(anb)
plt.axis('off')

plt.subplot(132)
plt.imshow(bna)
plt.axis('off')

diff = cv2.subtract(anb,bna)

plt.subplot(133)
plt.imshow(diff)
plt.axis('off')
```

```python
# idempotent property with union
aua=cv2.max(im1,im1)

plt.subplot(131)
plt.imshow(aua)
plt.axis('off')

plt.subplot(132)
plt.imshow(im1)
plt.axis('off')

diff = cv2.subtract(aua,im1)

plt.subplot(133)
plt.imshow(diff)
plt.axis('off')
```

```python
# idempotent property with intersection
ana=cv2.min(im1,im1)

plt.subplot(131)
plt.imshow(ana)
plt.axis('off')

plt.subplot(132)
plt.imshow(im1)
plt.axis('off')

diff = cv2.subtract(ana,im1)

plt.subplot(133)
plt.imshow(diff)
plt.axis('off')
```
