```python
#importing the necessary packages
import cv2
import matplotlib.pyplot as plt
```

```python
# Question 1
#reading the images
image = cv2.imread("/home/dmacs/Desktop/IP_Test/text_image.jpg", cv2.IMREAD_GRAYSCALE)
logo = cv2.imread("/home/dmacs/Desktop/IP_Test/dip_logo.jpg", cv2.IMREAD_GRAYSCALE)

negative_logo = 255 - logo  # finding the negative of the image

top_logo = cv2.resize(negative_logo, (48, 36)) #resizing the negative logo to be put in the header

image[1:37, 458:506] = top_logo  # inserting the logo as a sub-image header on the top right

plt.imshow(image, cmap="gray") #plotting the image
plt.axis("off")
```

```python
# Question 2
#reading the images
image = cv2.imread("/home/dmacs/Desktop/IP_Test/text_image.jpg", cv2.IMREAD_GRAYSCALE)
logo = cv2.imread("/home/dmacs/Desktop/IP_Test/dip_logo.jpg", cv2.IMREAD_GRAYSCALE)

#rotating the image by 45 degrees
matrix = cv2.getRotationMatrix2D((30, 75), 45, 0.3)
diagonal_logo = cv2.warpAffine(logo, matrix, (128, 128)) 

negative_logo = 255 - diagonal_logo  # finding the negative of the logo
resized_logo = cv2.resize(negative_logo, (508, 444)) #resizing the logo  

watermarked_image = cv2.addWeighted(image, 0.6, resized_logo, 0.4, 0) #adding the two images

#plotting the image
plt.imshow(watermarked_image, cmap="gray")
plt.axis("off")
```
